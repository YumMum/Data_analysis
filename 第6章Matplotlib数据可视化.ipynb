```python
# 折线图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib inline 是 Jupyter Notebook 中的一个魔法命令，用于在 Notebook 环境中直接显示 matplotlib 绘制的图形。它的作用是将图形嵌入到当前的单元格输出中，而不是弹出独立的窗口。
plt.rcParams["font.family"] = ["SimHei"]
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,marker='*',linewidth = 1,linestyle = '--',color = 'orange')
plt.plot(x,z)
# 设置标题和坐标轴标签
plt.title('正弦与余弦函数', fontsize=16)
plt.xlabel('角度 (弧度)', fontsize=14)
plt.ylabel('函数值', fontsize=14)
plt.legend(['sin','cos'],loc = 'best') # legend:图例
plt.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots()
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False
x1=np.arange(1,30)
y1=np.sin(x1)
ax1=plt.subplot(1,1,1)
plt.title('散点图')
plt.xlabel('X')
plt.ylabel("Y")
lvalue=x1
ax1.scatter(x1,y1,c='b',s=100,linewidth=lvalue,marker='o')
plt.legend('x1',loc='lower left')
plt.show()
```

```python
import pandas as pd
fig,axes=plt.subplots(nrows=2)
data=pd.Series(np.random.randn(16),index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
data.plot.barh(ax=axes[1],color='k',alpha=0.7)
```

```python
fig,ax=plt.subplots()
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False
x=np.arange(1,6)
Y1=np.random.uniform(1.5,1.0,5)
Y2=np.random.uniform(1.5,1.0,5)
plt.bar(x,Y1,width=0.35,facecolor='lightskyblue',edgecolor='white')
plt.bar(x+0.35,Y2,width=0.35,facecolor='yellowgreen',edgecolor='white')
plt.legend(['Y1','Y2'],loc='best')
plt.show()
```

```python
plt.figure(figsize=(6,6))
labels=['Spring','Summer','Autumn','Winter']
x=[15,30,45,10]
explode=[0.05,0.05,0.05,0.25]
plt.pie(x,labels=labels,explode=explode,startangle=60,autopct='%1.1f%%')
plt.title('Rainy days by season')
plt.legend(['Spring','Summer','Autumn','Winter'],loc='best')
plt.show()
```
